# Apply Transformations at data

```python
scaler = StandardScaler()
```

```python
X=cement.drop(columns = 'Strength' , axis = 1)
y = cement.iloc[:,-1]
```

```python
X = scaler.fit_transform(X)
```

```python
X_train , X_test ,y_train , y_test = train_test_split(X, y , train_size = 0.8 , random_state = 42)
```

```python
lr = LinearRegression()
lr.fit(X_train , y_train)
y_pred = lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# Cross checking with cross val score
lr = LinearRegression()
np.round(np.mean(cross_val_score(lr,X,y,scoring='r2')) * 100 , 0)
```

```python
# # Plotting Displot without applying transformations 
# for col in X_train.column:
#     plt.figure(figsize=(14,4))
#     sns.displot(df[col])
#     plt.title(col)
#     plt.show()
for i in range(X_train.shape[1]):   # number of columns
    sns.displot(X_train[:, i])      # take column by index
    plt.title(f"Feature {i}")
    plt.show()
```

```python
# Apply Box Cox Transformation at data 

# pt = PowerTransformer('box-cox')
# X_train_trans = pt.fit_transform(X_train + 0.000001)
# X_test_trans = pt.fit_transform(X_test + 0.000001)

# Applying Box-Cox Transform

# pt = PowerTransformer(method='box-cox')
# not apply becuase dataset having -ive values

# Applying Yeo_johnson Transform
pt = PowerTransformer(method='yeo-johnson')   # supports negatives and zeros

X_train_transformed = pt.fit_transform(X_train+0.000001)
X_test_transformed = pt.transform(X_test+0.000001)
```

```python
lr.fit(X_train_transformed , y_train)
x_pre = lr.predict(X_test_transformed)
r2_score(x_pre , y_test)
```

```python
# Apply Yeo-Johnson transform

pt1 = PowerTransformer()

X_train_transformed2 = pt1.fit_transform(X_train)
X_test_transformed2 = pt1.transform(X_test)

lr = LinearRegression()
lr.fit(X_train_transformed2,y_train)

y_pred3 = lr.predict(X_test_transformed2)

print(r2_score(y_test,y_pred3))
```
