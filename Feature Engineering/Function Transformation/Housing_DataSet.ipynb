```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# -----------------------
# Load dataset
# -----------------------
df = pd.read_csv("housing.csv")   # replace with your path

# Features & Target
X = df.drop('median_house_value', axis=1)
y = df['median_house_value']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)

# -----------------------
# Preprocessing
# -----------------------
num_cols = X.drop(columns=['ocean_proximity']).columns
cat_cols = ['ocean_proximity']

# Columns that are skewed
skewed_cols = ['total_rooms', 'total_bedrooms', 'population', 'households']
normal_cols = [col for col in num_cols if col not in skewed_cols]

# ColumnTransformer with Imputer + Scaling + Encoding
preprocessor = ColumnTransformer(
    transformers=[
        # Skewed numeric  impute + log + scale
        ('num_skew', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),
            ('log', FunctionTransformer(np.log1p, validate=False)),
            ('scaler', StandardScaler())
        ]), skewed_cols),

        # Normal numeric  impute + scale
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ]), normal_cols),

        # Categorical  impute + one-hot
        ('cat', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('ohe', OneHotEncoder(drop='first', sparse_output=False))
        ]), cat_cols)
    ],
    remainder='passthrough'
)

# -----------------------
# Pipelines
# -----------------------
# Linear Regression
lr_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('lr', LinearRegression())
])

# Random Forest
rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('rf', RandomForestRegressor(n_estimators=200, random_state=0))
])

# -----------------------
# Training
# -----------------------
lr_pipeline.fit(X_train, y_train)
rf_pipeline.fit(X_train, y_train)

# Predictions
y_pred_lr = lr_pipeline.predict(X_test)
y_pred_rf = rf_pipeline.predict(X_test)

# -----------------------
# Evaluation
# -----------------------
# print("Linear Regression")
# print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lr)))
# print("R Score:", r2_score(y_test, y_pred_lr))

# print("\nRandom Forest")
# print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
# print("R Score:", r2_score(y_test, y_pred_rf)))
```
