```python

# Crosstab: odor vs gill-size with class counts
heatmap_data = pd.crosstab([df['odor'], df['gill-size']], df['class'])

# Plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu")

plt.title("Heatmap of Odor & Gill-Size vs Class (Edible/Poisonous)")
plt.ylabel("Odor - Gill-Size")
plt.xlabel("Class (Edible/Poisonous)")
plt.show()
```

# Step 3 
# Encoding 

```python
# Remove Redantant Columns 
df.drop(['veil-type', 'veil-color' , 'ring-number' , 'stalk-shape' , 'gill-attachment'], axis = 1 , inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder , LabelEncoder , OrdinalEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train , X_test , y_train , y_test = train_test_split(df.iloc[:,1:] , df.iloc[: , 0] ,test_size = 0.3,  random_state = 0 )
```

```python
le = LabelEncoder()
ohe = OneHotEncoder()
oe = OrdinalEncoder()
```

```python
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
ordinal_col = ['gill-size']
nominal_col = X_train.drop(columns = ordinal_col , axis= 1 ).columns
```

```python
transformer = ColumnTransformer(
    transformers = [
        ('ohe' , OneHotEncoder(drop = 'first' , sparse_output = False) ,nominal_col ),
        ('oe' , OrdinalEncoder(categories=[['b' ,'n']]) , ordinal_col)
    ]
)
```

```python
X_train
```

```python
X_train_enc = transformer.fit_transform(X_train)
X_test_enc = transformer.transform(X_test)
```
