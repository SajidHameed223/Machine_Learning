```python
import numpy as np 
import  pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
wine = pd.read_csv('wine_data.csv')
```

```python
wine=wine.iloc[:, 1:5]
```

```python
wine.columns = ['Alcohol'  , 'Acidity' ,  'Malic Acid' , 'Sulfur']
```

```python
from sklearn.preprocessing import StandardScaler , MinMaxScaler , RobustScaler
```

```python
scaler = StandardScaler()
stand_wine = scaler.fit_transform(wine)
stand_wine = pd.DataFrame(stand_wine , columns = wine.column)
```

```python
scaler = MinMaxScaler()
minmax_wine = scaler.fit_transform(wine)
minmax_wine = pd.DataFrame(minmax_wine , columns = wine.column)
```

```python
scaler = RobustScaler()
robust_wine = scaler.fit_transform(wine)
robust_wine = pd.DataFrame(robust_wine , columns = wine.column)
```

```python
plt.figure(figsize = (15,4))
plt.subplot(1,4,1)
sns.kdeplot(data = wine , x = 'Alcohol' , c='red')
sns.kdeplot(data = wine , x = 'Acidity' , c='green')
sns.kdeplot(data = wine , x = 'Malic Acid', c ='blue')
sns.kdeplot(data = wine , x = 'Sulfur' , c='yellow')
plt.title('Without Scaling')

plt.subplot(1,4,2)
sns.kdeplot(data = stand_wine , x = 'Alcohol' , c='red')
sns.kdeplot(data = stand_wine , x = 'Acidity' , c='green')
sns.kdeplot(data = stand_wine , x = 'Malic Acid', c ='blue')
sns.kdeplot(data = stand_wine , x = 'Sulfur' , c='yellow')
plt.title('After Standardization')

plt.subplot(1,4,3)
sns.kdeplot(data = minmax_wine , x = 'Alcohol' , c='red')
sns.kdeplot(data = minmax_wine , x = 'Acidity' , c='green')
sns.kdeplot(data = minmax_wine , x = 'Malic Acid', c ='blue')
sns.kdeplot(data = minmax_wine , x = 'Sulfur' , c='yellow')
plt.title('After Min Max Normalization')

plt.subplot(1,4,4)
sns.kdeplot(data = robust_wine , x = 'Alcohol' , c='red')
sns.kdeplot(data = robust_wine , x = 'Acidity' , c='green')
sns.kdeplot(data = robust_wine , x = 'Malic Acid', c ='blue')
sns.kdeplot(data = robust_wine , x = 'Sulfur' , c='yellow')
plt.title('After Robust Normalization')
```

```python
plt.figure(figsize=(16,4))

plt.subplot(141)
sns.scatterplot(data=wine ,x= 'Alcohol' ,y= 'Sulfur')
plt.title('Before Scaling')


plt.subplot(142)
sns.scatterplot(data=stand_wine ,x= 'Alcohol' ,y= 'Sulfur')
plt.title('After Standard Scaling')


plt.subplot(143)
sns.scatterplot(data=minmax_wine ,x= 'Alcohol' ,y= 'Sulfur')
plt.title('After Min Max Scaling')


plt.subplot(144)
sns.scatterplot(data=robust_wine ,x= 'Alcohol' ,y= 'Sulfur')
plt.title('After Robust Scaling')

```
