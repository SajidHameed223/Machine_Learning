# EDA Step 2: Uni , Bi and MultiVarient Analysis?

## UniVarient Analysis

```python
df['2020'].hist(bins= 10)
```

```python
sns.boxplot(df['2020'])
```

```python
sns.displot(x='2020' , data = df , kde= True)
```

```python
# df.nlargest(10 , '2020').plot(kind='bar' )
df.nlargest(10, '2025').plot(x='Country', y='2020', kind='bar')
```
