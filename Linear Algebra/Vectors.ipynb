```python
import matplotlib.pyplot as plt
```

```python
a=[1,2,3,4,5]
b=[6,7,8,9,10]
# For one vector
# plt.scatter(x = 5, y=4)
# for a multiple vectors
plt.scatter(x=a,y=b)
plt.axis([0,6,0,12])
plt.show()
```

```python
# for vector space 
vectors= [(1,2),(4,5),(6,1),(3,2)]

for v in vectors:
    plt.arrow(0,0,v[0],v[1], head_length= 0.2 , head_width=0.3)
plt.xlim(0,8)
plt.ylim(0,6)
plt.show()
```

```python
# more friendly then and customizeable than arrows is quiver method 
# lets assume a vectors array 
vectors = [(5,1) , (4,2),(4,4),(-1,4),(-3,-4) ,(4,-6)]
for v in vectors:
    plt.quiver(0,0,v[0],v[1] , angles = 'xy' , scale_units = 'xy' , scale = 1)
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.axhline(0,linewidth=1)
plt.axvline(0,linewidth=1)
plt.show()    
```

```python
# Scalar Multiplication of vectors 
import numpy as np
vector = np.array([1,2])
vector = 4*vector
plt.arrow(0,0,vector[0],vector[1] ,head_width=0.2 ,head_length = 0.2)
plt.xlim(-6,9)
plt.ylim(-6,9)
plt.axhline(0,linewidth=1)
plt.axvline(0,linewidth=1)
plt.show()
```

```python
# Vector addiction 
v1 = np.array([1,4])
v2 = np.array([2,3])
v3 = v1+v2
plt.quiver(0,0,v3[0],v3[1]  , angles = 'xy' , scale_units='xy', scale =1)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```
