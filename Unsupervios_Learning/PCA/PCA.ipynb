```python
# Data Loading
import pandas as pd
wine_data_path = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQDVwxneOKOaJL13QMhkAhYrgWlH1tICY7RacUnj_lL8m9uUWaaUf3p7bScNyh_D2Rvt7nc1q11adSy/pub?gid=647503637&single=true&output=csv"
wine = pd.read_csv(wine_data_path)
wine.head()
```

# Fill Null Values

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer()
```

```python
X = wine.iloc[: , 1:]
y = wine.iloc[: , 0]
```

```python
X_trf = imputer.fit_transform(X)
```

```python
X_trf  = pd.DataFrame(X_trf , columns = X.columns)
```

```python
X_trf.sample(6)
```

```python
from sklearn.preprocessing import StandardScaler , LabelEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
```

```python
X_train , X_test , y_train , y_test = train_test_split(X_trf , y , test_size=0.2 , random_state=42)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
le = LabelEncoder()
```

```python
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train , y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test , y_pred))
```

# With PCA

```python
pca = PCA(n_components = 5 )
```

```python
pca_train = pca.fit_transform(X_train)
pca_test = pca.transform(X_test)
```

```python
# The principal components
components = pca.components_
# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
# Print the results
print("\nPrincipal Components:")
print(components)
print("\nExplained Variance Ratio:")
print(explained_variance_ratio)
```

```python
dt.fit(pca_train , y_train)
```

```python
y_pred = dt.predict(pca_test)
```

```python
print(accuracy_score(y_test , y_pred))
```
